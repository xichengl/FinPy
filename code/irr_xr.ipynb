```python
from _config import *

import numpy as np
import pandas as pd
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib import colormaps
tab20 = colormaps['tab20']
```

```python
cf1 = np.array([-100, 20, 30, 40, 50])
cf2 = np.array([100, -20, -40, -50, -10])
cf3 = np.array([-180, 40, 50, 40, 100])
r = np.linspace(0.01, 0.15, 100)
df = np.array([[1/(1+rate)**i for i in range(0,5)] for rate in r])
npv1 = df @ cf1
npv2 = df @ cf2
npv3 = df @ cf3
df_plot = pd.DataFrame({'r': r, 'y1': npv1, 'y2': npv2, 'y3': npv3})
```

```python
f, ax = plt.subplots(figsize = (10, 6))

sns.lineplot(data = df_plot, x = 'r', y = 'y3', ax = ax, label = 'Project 1', color = c1)
sns.lineplot(data = df_plot, x = 'r', y = 'y1', ax = ax, label = 'Project 2', color = c0)

ax.set_xlim(0.01, 0.15)
ax.set_ylim(-30, 50)
ax.hlines(0, 0.01, 0.15, linestyle = '--', linewidth = 0.8, color = 'black', alpha = 0.5)
ax.vlines(0.042, 0, 24.5, linestyle = '--', linewidth = 0.8, color = 'black', alpha = 0.5)

ax.annotate(r'$xr$', xy = (0.04,0), xytext = (0.041, -3), color = 'black', fontsize = 12)
ax.annotate(r'$irr_1$', xy = (0.09,0), xytext = (0.087, -3), color = 'black', fontsize = 12)
ax.annotate(r'$irr_2$', xy = (0.125,0), xytext = (0.127, -3), color = 'black', fontsize = 12)

ax.set_xlabel('Hurdle Rate', fontweight = 'bold', fontsize = 12, labelpad = 10)
ax.set_ylabel('NPV', fontweight = 'bold', fontsize = 12)
# ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.show()
f.savefig('figures/cross_over_rate.png', format='png', dpi=800, bbox_inches='tight')
```
